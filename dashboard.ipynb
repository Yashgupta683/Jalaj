```python
# Install necessary libraries
!pip install dash plotly pandas numpy pyngrok flask

import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import pandas as pd
import numpy as np
from flask import Flask
from pyngrok import ngrok

# Set up ngrok authentication
NGROK_AUTH_TOKEN = "2uiGY2H4X4s8skbkyMgdgZ0QpES_5y3pEr6U5XKF4MdnDV1me"  # Replace with your ngrok authtoken
ngrok.set_auth_token(NGROK_AUTH_TOKEN)

# Create a Flask server
server = Flask(__name__)
app = dash.Dash(__name__, server=server)

# Start ngrok tunnel
public_url = ngrok.connect(5000).public_url
print(f"Dashboard is live at: {public_url}")

# Generate sample data
np.random.seed(42)
data = pd.DataFrame({
    "Time": pd.date_range(start="2024-01-01", periods=50, freq="min"),
    "Temperature": np.random.uniform(20, 35, 50),
    "Humidity": np.random.uniform(30, 70, 50)
})

# App layout
app.layout = html.Div([
    html.H1("Real-time Temperature and Humidity Dashboard"),
    dcc.Graph(id='line-chart'),
    dcc.Interval(id='interval-component', interval=2000, n_intervals=0),
])

# Callback to update the graph
@app.callback(
    Output('line-chart', 'figure'),
    Input('interval-component', 'n_intervals')
)
def update_graph(n):
    global data
    new_time = data.iloc[-1]['Time'] + pd.Timedelta(minutes=1)
    new_temp = np.random.uniform(20, 35)
    new_humidity = np.random.uniform(30, 70)
    new_data = pd.DataFrame({"Time": [new_time], "Temperature": [new_temp], "Humidity": [new_humidity]})
    data = pd.concat([data, new_data]).iloc[1:]

    fig = go.Figure()
    fig.add_trace(go.Scatter(x=data['Time'], y=data['Temperature'], mode='lines', name='Temperature', line=dict(color='red')))
    fig.add_trace(go.Scatter(x=data['Time'], y=data['Humidity'], mode='lines', name='Humidity', line=dict(color='blue')))
    fig.update_layout(title='Temperature & Humidity Trends', xaxis_title='Time', yaxis_title='Values', template='plotly_dark')
    return fig

# Run the server
if __name__ == '__main__':
    server.run(port=5000)
```
